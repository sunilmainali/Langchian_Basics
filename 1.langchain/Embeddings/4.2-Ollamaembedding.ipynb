#### Ollama
Ollama supports embedding models, making it possible to build retrieval augmented generation (RAG) applications that combine text prompts with existing documents or other data.

```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings=(OllamaEmbeddings(model="nomic-embed-text")) ## default llama2
```

```python
embeddings
```

```python
r1=embeddings.embed_query("Alpha is the first letter of Greek alphabet")
r1
```

```python
from langchain_community.embeddings import OllamaEmbeddings

embeddings = OllamaEmbeddings(model=" nomic-embed-text")  # or just "llama2" as default

r1 = embeddings.embed_query(
    "Alpha is the first letter of Greek alphabet"
)
print(r1[:5])  # prints first 5 values of the embedding vector
```
